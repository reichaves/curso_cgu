```python
# Renomeia colunas para nomes legiveis
```

```python
pep_sp.rename(columns = {'Sigla_FunÃ§Ã£o':'sigla_funcao', 'DescriÃ§Ã£o_FunÃ§Ã£o':'descricao_funcao', 'NÃ­vel_FunÃ§Ã£o':'nivel_funcao', 'Nome_ÃrgÃ£o':'nome_orgao', 'Data_InÃ­cio_ExercÃ­cio':'data_inicio_exercicio', 'Data_Fim_ExercÃ­cio':'data_fim_exercicio', 'Data_Fim_CarÃªncia':'data_fim_carencia' 
                              }, inplace = True) 
```

```python
# Tira pontuacoes do CPF
```

```python
pep_sp['CPF'] = pep_sp['CPF'].str.replace('.','')
pep_sp['CPF'] = pep_sp['CPF'].str.replace('-','')
```

```python
# Cria dataframe das PEPs que sao socias de empresas - compara CPF e nome ao mesmo tempo
```

```python
empresas_pep_sp = pd.merge(pep_sp, socios, left_on=['CPF', 'Nome_PEP'], right_on=['cnpj_cpf_socio', 'nome_socio'])
```

```python
empresas_pep_sp.info()
```

```python
# Verifica quais dessas empresas receberam pagamentos
```

```python
empresas_pep_sp_recebeu_recursos = pd.merge(empresas_pep_sp, despesas, left_on='cnpj', right_on='nr_identificador_despesa')
```

```python
empresas_pep_sp_recebeu_recursos.info()
```

```python
empresas_pep_sp_recebeu_recursos.to_excel("resultados/empresas_pep_sp_recebeu_recursos.xlsx", index=False, sheet_name='Sheet1')  
```

```python
# Ordena os pagamentos
```

```python
empresas_pep_sp_recebeu_recursos.groupby(['ds_despesa', 'Nome_PEP'], sort = False).sum()
```

```python
# Cria coluna do municipio onde esta a PEP
```

```python
# Pega primeiro valor em nome_orgao
```

```python
empresas_pep_sp_recebeu_recursos['municipio_da_pep'] = empresas_pep_sp_recebeu_recursos['nome_orgao'].apply(lambda t: t.split('-')[0].strip())
```

```python
empresas_pep_sp_recebeu_recursos['municipio_da_pep'].head()
```

```python
# Retira so o conteudo do item 3 para tras
```

```python
def limpa1(linha):
    texto = linha['municipio_da_pep'].split(' ')
    cidade = texto[2:]
    
    cidade = " ".join(str(x) for x in cidade)
    
    return cidade
```

```python
empresas_pep_sp_recebeu_recursos['municipio_da_pep'] = empresas_pep_sp_recebeu_recursos.apply(limpa1, axis=1)
```

```python
empresas_pep_sp_recebeu_recursos.head()
```

```python
# Apenas com as empresas de pessoas politicamente expostas que receberam recursos dos próprios municípios em que tem mandato
```

```python
empresas_pep_sp_recebeu_recursos['compara_cidades'] = empresas_pep_sp_recebeu_recursos.apply(lambda x : "sim" if x['ds_municipio'] == x['municipio_da_pep'] else "nao", axis=1)
```

```python
empresas_pep_sp_recebeu_recursos.head()
```

```python
empresas_pep_sp_recebeu_recursos_mesmo_municipio = empresas_pep_sp_recebeu_recursos[(empresas_pep_sp_recebeu_recursos['compara_cidades'] == 'sim')]
```

```python
empresas_pep_sp_recebeu_recursos_mesmo_municipio.info()
```
